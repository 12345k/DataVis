```python
winloss = []
for t in df.team1.unique():
    for s in df.season.unique():
        won = len(df[(df.winner == t) & (df.season == s)])
        played = len(df[((df.team1 == t) | (df.team2 == t)) & (df.season == s)])
        if((won != 0) & (played != 0)):
            winloss.append(won/played)
        else:
            winloss.append(0)
        
        
winloss_ser = pd.Series(winloss, 
                        index=pd.MultiIndex.from_product([df.team1.unique(),df.season.unique()]))

```

```python
winloss_types = ['winloss','winlossTosswon','winlossBat','winlossField'] #, 'winlossHome', 'winlossAway']

winloss = []
for t in df.team1.unique():
    for wl in winloss_types:
        won, loss = 0, 0
        if wl == 'winloss':
            won = len(df[df.winner == t])
            played = len(df[(df.team1 == t) | (df.team2 == t)])
        elif wl == 'winlossTosswon':
            won = len(df[(df.winner == df.toss_winner) & (df.winner == t)])
            played = len(df[((df.team1 == t) | (df.team2 == t)) & (df.winner == df.toss_winner)])
        elif wl == 'winlossBat':
            won = len(df[(df.winner == t) & 
               (((df.toss_winner == t) & (df.toss_decision == 'bat'))
                   | ((df.toss_winner != t) & (df.toss_decision == 'field')))])
            played = len(df[((df.team1 == t) | (df.team2 == t)) &
                          (((df.toss_winner == t) & (df.toss_decision == 'bat'))
                           | ((df.toss_winner != t) & (df.toss_decision == 'field')))])
        elif wl == 'winlossField':
            won = len(df[(df.winner == t) & 
               (((df.toss_winner == t) & (df.toss_decision == 'field'))
                   | ((df.toss_winner != t) & (df.toss_decision == 'bat')))])
            played = len(df[((df.team1 == t) | (df.team2 == t)) &
                          (((df.toss_winner == t) & (df.toss_decision == 'field'))
                           | ((df.toss_winner != t) & (df.toss_decision == 'bat')))])
        
        if((won != 0) & (played != 0)):
             winloss.append(np.round(won/played,2))
        else:
            winloss.append(0)
        #print(t,wl, won/played)
       
                
```

```python
winloss_ser2 = pd.Series(winloss, index=pd.MultiIndex.from_product([df.team1.unique(), winloss_types]))
print(winloss_ser2)
print(winloss_ser2.unstack())
```
